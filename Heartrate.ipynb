```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Create a DataFrame with the data
data = {
    "Mammal": ["Vespergo pipistrellas", "Mouse", "Rat", "Guinea Pig", "Rabbit", "Little Dog", "Big Dog", "Sheep", "Man", "Horse", "Ox", "Elephant"],
    "Body Weight (kg)": [0.004, 0.025, 0.2, 0.3, 2, 5, 30, 50, 70, 450, 500, 3000],
    "Pulse Rate (beats/min)": [660, 670, 420, 300, 205, 120, 85, 70, 72, 38, 40, 48]
}
df = pd.DataFrame(data)

# Calculate predicted heart rates using the model equation
k = 100  # Adjust this constant for scaling
df["Predicted HR (beats/min)"] = k / df["Body Weight (kg)"]**(1/3)

# Plot actual vs. predicted heart rates
plt.figure(figsize=(8, 6))
plt.scatter(df["Body Weight (kg)"], df["Pulse Rate (beats/min)"], label="Actual HR")
plt.scatter(df["Body Weight (kg)"], df["Predicted HR (beats/min)"], label="Predicted HR")
plt.xlabel("Body Weight (kg)")
plt.ylabel("Heart Rate (beats/min)")
plt.title("Actual vs. Predicted Heart Rate")
plt.legend()
plt.grid(True)

# Calculate mean squared error
mse = mean_squared_error(df["Pulse Rate (beats/min)"], df["Predicted HR (beats/min)"])
print("Mean Squared Error:", mse)

plt.show()
```
